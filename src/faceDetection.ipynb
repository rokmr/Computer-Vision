# Haar Cascade

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the Haar Cascade classifiers from OpenCV's data directory
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_detector = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def adjusted_detect_face(img):
    face_img = img.copy()
    face_rect = face_detector.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)

    for (x, y, w, h) in face_rect:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (0, 255, 255), 3)
        
    return face_img

def detect_eyes(img):
    eye_img = img.copy()
    eye_rect = eye_detector.detectMultiScale(eye_img, scaleFactor=1.2, minNeighbors=5)

    for (x, y, w, h) in eye_rect:
        cv2.rectangle(eye_img, (x, y), (x + w, y + h), (0, 255, 255), 3)
        
    return eye_img
```

```python
img = cv2.imread('../data/rohit.png')
img_copy1 = img.copy()


plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
face = adjusted_detect_face(img_copy1)
face_eyes = detect_eyes(face)
plt.imshow(cv2.cvtColor(face_eyes, cv2.COLOR_BGR2RGB))
plt.show()
```

# HOG + SVM

```python
import dlib
import cv2

# Initialize Dlib face detector
detector = dlib.get_frontal_face_detector()

# Load image
img = cv2.imread('../data/rohit.png')

# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = detector(gray)

# Draw rectangles around faces
for face in faces:
    x, y, w, h = (face.left(), face.top(), face.width(), face.height())
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

plt.imshow(img[..., ::-1])
```
